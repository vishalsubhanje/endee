```python
df_clean = df[[
    "Name",
    "Sex",
    "Equipment",
    "Age",
    "BodyweightKg",
    "BestSquatKg",
    "BestBenchKg",
    "BestDeadliftKg",
    "TotalKg"
]].dropna()

def row_to_text(row):
    return (
        f"Lifter {row['Name']} is a {row['Sex']} athlete using {row['Equipment']} equipment. "
        f"Age: {row['Age']}. Bodyweight: {row['BodyweightKg']} kg. "
        f"Best squat: {row['BestSquatKg']} kg. "
        f"Best bench press: {row['BestBenchKg']} kg. "
        f"Best deadlift: {row['BestDeadliftKg']} kg. "
        f"Total lifted: {row['TotalKg']} kg."
    )

df_clean["text"] = df_clean.apply(row_to_text, axis=1)
df_sample = df_clean.sample(1000, random_state=42).reset_index(drop=True)

print(len(df_sample))
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(df_sample["text"].tolist())

print(len(embeddings))
```

```python
df_sample["embedding"] = list(embeddings)

df_sample[["Name", "text", "embedding"]].head()
```

```python
records = []

for i, row in df_sample.iterrows():
    records.append({
        "id": str(i),
        "text": row["text"],
        "embedding": row["embedding"],
        "metadata": {
            "name": row["Name"],
            "sex": row["Sex"],
            "equipment": row["Equipment"],
            "age": float(row["Age"]),
            "bodyweight": float(row["BodyweightKg"]),
            "best_squat": float(row["BestSquatKg"]),
            "best_bench": float(row["BestBenchKg"]),
            "best_deadlift": float(row["BestDeadliftKg"]),
            "total": float(row["TotalKg"])
        }
    })

print(len(records))
print(records[0].keys())
```
